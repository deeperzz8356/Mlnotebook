A=LU decmposition

```python
import scipy.linalg as lu
A=np.array([
    [4,3],
    [6,3],
    [8,5]
])
P,L,U=lu.lu(A)
print("P:\n",P)
print("L:\n",L) 
print("U:\n",U)
print("Reconstructed A:\n",P@L@U)
```

Gram schmidt

```python
A=np.array([[4, 3], [6, 3]])
Q,R=np.linalg.qr(A)
print("Q:\n",Q)
print("R:\n",R)
print("Reconstructed A:\n",Q@R)
```

Least square

```python
A=np.array([[4, 3], [6, 3]])
b=np.array([10,12])

Q,R=lu.qr(A)
y=np.linalg.solve(R,Q.T@b)
print("Solution x:\n",y)
y_np,*_=np.linalg.lstsq(A,b,rcond=None)
print("Numpy lstsq Solution x:\n",y_np)
```

Eigen value

```python
#eigen values and vectors
A=np.array([[4, -2], [1, 1]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:\n",eigenvalues)
print("Eigenvectors:\n",eigenvectors)
print("A v:",A @ eigenvectors[:,0])
print("λ v:",eigenvalues[0] * eigenvectors[:,0])
```

Symettric matric eigen value

```python
#eigen value and vector of symmetric matrix
A=np.array([[4, 2], [2, 3]])
eigenvalues, eigenvectors = np.linalg.eigh(A)
print("Eigenvalues:\n",eigenvalues) 
print("Eigenvectors:\n",eigenvectors)
```

Positive definite

```python
def positie_definite(A):
    try:
        np.linalg.cholesky(A)
        return True
    except np.linalg.LinAlgError:
        return False
    
A=np.array([[4, 2], [2, 3]])
e,ev=np.linalg.eig(A)
print("Eigenvalues:",e)
if np.all(e>0):
    print("All eigenvalues are positive.")  
print("Is A positive definite?",positie_definite(A))
```

Single value decompostion

```python
#svd and rank ,single values,null space
A=np.array([[4, 3], [6, 3]])
U,S,Vt=np.linalg.svd(A)
print("U:\n",U)
print("S:\n",S)
print("Vt:\n",Vt)
print("Reconstructed A:\n",U @ np.diag(S) @ Vt)
print("Rank of A:",np.linalg.matrix_rank(A))
print("Null space of A:",Vt.T[:,2-1:])  #last column of Vt transposed
```
